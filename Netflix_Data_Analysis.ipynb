## Data Cleaning

```python
# Remove duplicate rows
df = df.drop_duplicates()

# Confirm duplicates removed
df.duplicated().sum()
```

```python
# Remove rows with missing values
df = df.dropna()

# Confirm missing values removed
df.isnull().sum()
```

## Data Preprocessing / Data Transformation

```python
# Convert Release_Date to datetime format
df['Release_Date'] = pd.to_datetime(df['Release_Date'])

# Extract year from release date
df['Release_Date'] = df['Release_Date'].dt.year

# Display sample row
df.head(1)
```

```python
# Drop columns not needed for analysis
df = df.drop(columns=['Overview','Poster_Url','Original_Language'])
```

```python
# Convert genre string into list
df['Genre'] = df['Genre'].str.split(', ')

# Split genres into multiple rows
df = df.explode('Genre').reset_index(drop=True)

# Show first rows
df.head()
```

## Column / Feature Analysis

```python
# Count frequency of each genre
df['Genre'].value_counts()
```

*  The Release_Date column was converted to datetime format, and the year was extracted to support time-based analysis of movie releases.
*   Columns such as Overview, Poster_Url, and Original_Language were removed as they were not relevant for the intended numerical and categorical analysis.

## Data Visualization

###1️⃣ Which genre appears most often?


```python
# Count how many times each genre appears in the dataset
genre_counts = df['Genre'].value_counts()

# Create a bar chart showing the frequency of each genre
fig = px.bar(
    x=genre_counts.index,      # Genre names on X-axis
    y=genre_counts.values,     # Number of movies in each genre
    title="Most Common Movie Genres"
)

# Display the chart
fig.show()
```

Conclusion:
Drama is the most common genre in the dataset.

###2️⃣ Which movies have the highest popularity?

```python
# Sort movies by popularity in descending order and select top 10
top_movies = df.drop_duplicates(subset="Title") \
               .sort_values(by="Popularity", ascending=False) \
               .head(10)

# Create a horizontal bar chart for better readability of movie titles
fig = px.bar(
    top_movies,
    x="Popularity",        # Popularity score on X-axis
    y="Title",             # Movie titles on Y-axis
    orientation="h",       # Horizontal bars
    title="Top 10 Most Popular Movies"
)

# Display the chart
fig.show()
```

Conclusion:
Spider-Man: No Way Home has the highest popularity score in the dataset.
Other highly popular movies include The Batman, No Exit and Encanto.

###3️⃣ How has movie production changed over time?

```python
# Remove duplicate movies first (since each movie appears multiple times for different genres)
unique_movies = df.drop_duplicates(subset='Title')

# Count how many unique movies were released each year
movies_per_year = unique_movies['Release_Date'].value_counts().sort_index()

# Create a line chart to show movie production trend over time
fig = px.line(
    x=movies_per_year.index,    # Years
    y=movies_per_year.values,   # Number of movies released
    markers=True,               # Show points on the line
    title="Movies Released Over the Years"
)

# Display the chart
fig.show()
```

Conclusion:
The number of movies released has increased steadily over time, especially after the year 2000.
The slight drop after 2020 may be due to incomplete data for recent years or because newer movies are still being added to the dataset.

###4️⃣ Which genres have the highest average popularity?

```python
# Calculate the average popularity score for each genre
genre_popularity = df.groupby('Genre')['Popularity'].mean().reset_index()

# Create a bar chart showing average popularity of each genre
fig = px.bar(
    genre_popularity,
    x="Genre",        # Genre names
    y="Popularity",   # Average popularity score
    title="Average Popularity by Genre"
)

# Display the chart
fig.show()
```

Conclusion: Adventure movies have the highest average popularity,suggesting that audiences strongly engage with adventure-based content.
Other genres such as Action and Science Fiction also show relatively high popularity.

###5️⃣ What is the distribution of popularity scores?

```python
# Create a histogram to understand how popularity scores are distributed
fig = px.histogram(
    df,
    x="Popularity",      # Popularity values
    nbins=30,            # Number of bins in the histogram
    title="Distribution of Movie Popularity"
)

# Display the chart
fig.show()
```

Conclusion:
The popularity distribution is highly skewed.
Most movies are average.
Only a few movies become blockbuster hits.

Overall Insights

From the analysis of the Netflix movies dataset, several interesting trends can be observed.

• Drama is the most frequently occurring genre in the dataset, indicating that Netflix offers a large number of drama-based movies.

• Adventure movies show the highest average popularity, suggesting that audiences strongly engage with adventurous and exciting content.

• Genres such as Action and Science Fiction also receive high audience interest.

• The number of movie releases increased significantly after the year 2000, showing the rapid growth of the film industry and streaming platforms.
